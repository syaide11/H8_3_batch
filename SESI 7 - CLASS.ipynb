## Plotting

##### Kita harus menggunakan plot dengan benar, harus sesuai dengan informasi yang disajikan

### Line Plot (menunjukkan perubahan terhadap waktu)

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

# styling
mpl.style.use(['ggplot'])
```

```python
# plot garis untuk data imigran haiti
years = list(range(1980,2014))
haiti = df_can.loc['Haiti', years]
haiti
```

```python
# Haiti dalam bentuk kolom
haiti.plot();

plt.title("Imigration from Haiti");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");

plt.text(2000,6000, 'Gempa Haiti 2010');
```

##### Pertanyaan 1 : Bagaimana perbandingan jumlah imigran China dari tahun 1980 - 2013 ke Canada

```python
negara = ['China', 'India']
china_india = df_can.loc[negara, years]
china_india
```

```python
# Tukar baris jadi kolom dan kolom jadi baris
china_india.plot();
```

```python
# Transpose
CI_transposed = china_india.T
CI_transposed.head()
```

```python
CI_transposed.plot()
plt.title("Imigran dari China dan India");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
```

##### Pertanyaan 2 : Bandingkan tren dari 5 negara teratas yang menyumbang imigran ke Kanada

```python
top5_index = df_can['Total'].sort_values(ascending=False).head().index
top5_df = df_can.loc[top5_index, years]
top5_df = top5_df.T
```

```python
top5_df.plot(figsize=(25,10));
plt.title("Imigran dari 5 negara teratas");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
```

##### Area Plot

```python
top5_df.plot(kind='area',
             stacked=False,
             figsize=(25,10));
plt.title("Imigran dari 5 negara teratas");
plt.ylabel("Jumlah Imigran");
plt.xlabel("Tahun");
```

##### Histogram (tidak sama dengan barplot)

```python
df_2013 = df_can[2013]
df_2013
```

```python
# numpy
count, bin_edges = np.histogram(df_2013, bins=10)

print(count)
print(bin_edges)
```

```python
df_2013.plot(kind='hist',
             bins=10,
             xticks=bin_edges,
             figsize=(10,5));

plt.title("Histogram dari data Imigran 2013");
plt.xlabel("Jumlah Imigran");
plt.ylabel("Jumlah Negara");
```

#### Pie Chart

```python
df_continents = df_can.groupby('Continent').sum()
df_continents
```

```python
df_continents['Total'].plot(kind='pie',
                            figsize=(9,6),
                            autopct="%.2f%%",
                            startangle=45,
                            shadow=True);
```

```python
explode_list = [0.1,0,0,0,0.1,0.1]
df_continents['Total'].plot(kind='pie',
                            figsize=(15,6),
                            autopct="%.2f%%",
                            startangle=45,
                            shadow=True,
                            labels=None,
                            pctdistance=1.15,
                            explode=explode_list);

plt.legend(labels=df_continents.index);
plt.title("Data Imigrasi sesuai Continent, y=1.12");
plt.axis('equal')
```
